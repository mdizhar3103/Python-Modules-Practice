## Removing Outliers

```python
athletes = pd.read_csv('athletes.csv')
athletes.info()
```

```python
%matplotlib inline
athletes.plot.scatter(x='height', y='weight')
```

```python
heights = athletes['height']
heights.plot.box()
```

```python
q1 = heights.quantile(.25)
q3 = heights.quantile(.75)
iqr = q3 -q1
pmin = q1 - 1.5 * iqr
pmax = q3 + 1.5 * iqr
# new heights
# nwh = heights[heights.between(pmin, pmax)]
nwh = heights.where(heights.between(pmin, pmax))
# removing outliers with null values
```

```python
compare = pd.DataFrame({'before': heights, 'after': nwh})
compare.plot.box()
compare.describe()
```

```python
# applying outliers to original df

heights.where(heights.between(pmin, pmax), inplace=True)
```

```python
athletes.plot.scatter(x='height', y='weight')
```

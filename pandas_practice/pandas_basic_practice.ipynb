### Indexing with right way

- Using loc and iloc
- Labels vs Positions
- df.loc[row_indexer, col_indexer] (*Selecting by labels*)
- df.iloc[row_indexer, col_indexer] (*Selecting by position*)

```python
# selecting specific column and row value using loc
print(pickle_data.loc[1035, "artist"])
```

```python
# selecting specific column and row value using iloc
print(pickle_data.iloc[0,0])
```

```python
# selecting all columns
print(pickle_data.iloc[0, :])
```

```python
print(pickle_data.iloc[0:2, 0:2])
```

```python
# Performing operations on dataframe
# the operation will fail in case on incompatible datatype, missing values, etc

pickle_data["height"] * pickle_data["width"]
```

```python
# if you notice the dtype is object, thats why our operation failed

# Converting Datatype in Pandas
# this will fail if encunter NaN values

pd.to_numeric(pickle_data["width"])
```

```python
# forcing NaNs values to Not a Number as it treat it as string
pd.to_numeric(pickle_data["width"], errors="coerce")
```

```python
# updating the columns value with new correct values

pickle_data.loc[:, "width"] = pd.to_numeric(pickle_data["width"], 
                                            errors="coerce")
```

```python
# repaeting for height column as well to perform operations

pickle_data.loc[:, "height"] = pd.to_numeric(pickle_data["height"], 
                                            errors="coerce")
```

```python
# now perform the multiplication operation

pickle_data["height"] * pickle_data["width"]
```

```python
# Conditional Formatting

artist_counts = file_data["artist"].value_counts()
artist_counts.head()
```

```python
writer = pd.ExcelWriter("colors.xlsx", engine="xlsxwriter")
artist_counts.to_excel(writer, sheet_name="Artist Counts")
sheet = writer.sheets["Artist Counts"]
cells_range = "B2:B{}".format(len(artist_counts.index))
sheet.conditional_format(cells_range, {'type': '2_color_scale',
                                       'min_value': '10', 
                                       'min_type': 'precentile', 
                                       'max_value': '99', 
                                       'max_type': 'percentile'})
writer.save()
```

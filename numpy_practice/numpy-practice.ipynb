***Iterating over numpy array***

```python
a = np.arange(11) ** 2
```

```python
for i in a:
    print(i ** (1/2))
```

```python
students = np.array([['Alice', 'Bob', 'Cathy', 'Dorothy'],
                    [65, 78, 90, 81],
                    [71, 82, 79, 92]])
```

```python
for i in students:
    print("i = ", i)
```

```python
# accessing each element in array

for element in students.flatten():
    print(element)
```

```python
# using flattern option: order='F' - this will print first column value of each row

for element in students.flatten(order='F'):
    print(element)
```

```python
x = np.arange(12).reshape(3, 4)
x
```

```python
# iterating each element row-wise

for i in np.nditer(x):
    print(i)
```

```python
# iterating each element col-wise using order option

for i in np.nditer(x, order='F'):
    print(i)
```

```python
# Note: flatten - returns a result array in flatten order
#       nditer - allows jsut to iterate over elements  
```

```python
for i in np.nditer(x, order='F', flags = ["external_loop"]):
    print(i)
```

```python
# Note: nditer is read-only operation we can't assign value to it

for arr in np.nditer(x):
    arr[...] = arr * arr
```

```python
# to achive the read-write operation use the following option

for arr in np.nditer(x, op_flags=["readwrite"]):
    arr[...] = arr * arr
```

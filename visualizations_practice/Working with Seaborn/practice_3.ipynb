```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fb_data = pd.read_csv("datasets/dataset_Facebook.csv",
                     sep=r"\s*;\s*",
                     engine="python")
fb_data.head()
```

```python
f, ax = plt.subplots(figsize=(10, 5))
sns.distplot(fb_data["Lifetime Post Total Reach"], ax=ax)
```

```python
# default setting

sns.set()
```

```python
# adding ticks on axes

sns.set_style("ticks")

f, ax = plt.subplots(figsize=(10, 5))
sns.distplot(fb_data["Lifetime Post Total Reach"], ax=ax)
```

```python
# chaning grid for current figure only use 'with' keyword

with sns.axes_style("darkgrid"):
    f, ax = plt.subplots(figsize=(10, 5))
    sns.distplot(fb_data["Lifetime Post Total Reach"], ax=ax)
```

```python
# removing top and right side lines of figure (default setting)

f, ax = plt.subplots(figsize=(10, 5))
sns.distplot(fb_data["Lifetime Post Total Reach"], ax=ax)
sns.despine()
```

```python
f, ax = plt.subplots(figsize=(10, 5))
sns.distplot(fb_data["Lifetime Post Total Reach"], ax=ax)

sns.despine(left=True, right=False, bottom=True, top=False)
```

```python
# shifting the axes

f, ax = plt.subplots(figsize=(10, 5))
sns.distplot(fb_data["Lifetime Post Total Reach"], ax=ax)

sns.despine(offset=15, trim = True)
# 15 in pixels
```

## Color Palettes

```python
sns.set_palette("Greens_d")

f, ax = plt.subplots(figsize=(10, 5))
sns.distplot(fb_data["Lifetime Post Total Reach"], ax=ax)
```

```python
sns.set_style("whitegrid")

with sns.color_palette("Blues"):
    f, ax = plt.subplots(figsize=(10, 5))
    sns.distplot(fb_data["Lifetime Post Total Reach"], ax=ax)
```

```python
# resetting to default
sns.set()
```

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
# display current palette used by seaborn
```

```python
f, ax = plt.subplots(figsize=(10, 5))

sns.set_context("poster")

sns.kdeplot(fb_data["Total Interactions"], bw_adjust=0.04, label="bw_adjust: 0.04")
sns.kdeplot(fb_data["Total Interactions"], bw_adjust=1, label="bw_adjust: 1")
sns.kdeplot(fb_data["Total Interactions"], bw_adjust=2, label="bw_adjust: 2")
sns.kdeplot(fb_data["Total Interactions"], bw_adjust=5, label="bw_adjust: 5")
sns.kdeplot(fb_data["Total Interactions"], bw_adjust=7, label="bw_adjust: 7")
sns.kdeplot(fb_data["Total Interactions"], bw_adjust=9, label="bw_adjust: 9")

plt.xlim(-50, 800)
```

```python
sns.palplot(sns.color_palette("hls", 12))
# hls: hue, lightness, saturation
```

```python
# specifying hls range
sns.palplot(sns.hls_palette(12, l=0.2, s=0.5))
```

```python
sns.choose_colorbrewer_palette('qualitative')
```

```python
sns.pairplot(fb_data,
             vars=["like", "share", "comment"],
             hue="Post Month",
             palette="Paired")
```

```python
sns.palplot(sns.color_palette(["#9b59b6", "#3498db", "#95a5a6", "#e74c3c","#34495e","#2ecc71"]))
```

```python
# using seaborn identifies for rgb name

f, ax = plt.subplots(figsize=(10, 5))

sns.xkcd_rgb["pale red"]
sns.kdeplot(fb_data["Total Interactions"], bw_adjust=0.04, label="bw_adjust: 0.04")

sns.xkcd_rgb["medium green"]
sns.kdeplot(fb_data["Total Interactions"], bw_adjust=1, label="bw_adjust: 1")

sns.xkcd_rgb["denim blue"]
sns.kdeplot(fb_data["Total Interactions"], bw_adjust=2, label="bw_adjust: 2")

plt.xlim(-50, 800)
```

```python
sns.palplot(sns.color_palette("Blues"))
```

```python
# reverse blue
sns.palplot(sns.color_palette("Blues_r"))
```

```python
# dark blues
sns.palplot(sns.color_palette("Blues_d"))
```

```python
# light palatte color

sns.palplot(sns.light_palette("green"))
```

```python
f, ax = plt.subplots(figsize=(10, 5))

cmap = sns.light_palette("green", as_cmap=True, reverse=True)
sns.kdeplot(fb_data["Total Interactions"], fb_data["Lifetime Post Total Reach"],
           cmap=cmap)

plt.xlim(-100, 400)
plt.ylim(-5000, 15000)
```

```python
sns.palplot(sns.dark_palette("purple", reverse=True))
```

```python
sns.choose_colorbrewer_palette("sequential")
```

```python
sns.palplot(sns.color_palette('cubehelix', 12))
```

```python
sns.palplot(sns.cubehelix_palette(12))
```

```python
f, ax = plt.subplots(figsize=(10, 5))

cmap = sns.cubehelix_palette(dark=1, light=0, as_cmap=True, reverse=True)
sns.kdeplot(fb_data["Total Interactions"], fb_data["Lifetime Post Total Reach"],
           cmap=cmap, shade=True)

plt.xlim(-200, 400)
plt.ylim(-5000, 15000)
```

```python
# Diverging colors
sns.palplot(sns.color_palette("BrBG", 9))
```

```python
sns.palplot(sns.color_palette("coolwarm", 9))
```

```python
sns.choose_colorbrewer_palette("diverging")
```

```python
g = sns.PairGrid(fb_data,
                vars=["comment", "like", "share"],
                hue="Type",
                palette="coolwarm")

g.map_offdiag(plt.scatter)
g.map_diag(sns.kdeplot)
g.add_legend()
```

## Trellis Plots in Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fb_data = pd.read_csv('datasets/dataset_Facebook.csv',
                     sep=r'\s*;\s*',
                     engine='python')
fb_data.head()
```

```python
fb_data.describe()
```

```python
# 3 dimension to the data

g = sns.FacetGrid(fb_data, col="Type")
# we have 4 types of posts in data so we have 4 grid
```

```python
g = sns.FacetGrid(fb_data, col="Type")
g.map(plt.hist, "Total Interactions")
```

```python
# for better view put the limit on x and y

g = sns.FacetGrid(fb_data, col="Type")
g.map(plt.hist, "Total Interactions")
plt.xlim(-20, 3000)
plt.ylim(0, 30)
```

```python
# plotting bivariate data

g = sns.FacetGrid(fb_data, col="Type")
g.map(plt.scatter, "Lifetime Post Total Reach", "Lifetime Engaged Users")
```

```python
g = sns.FacetGrid(fb_data, 
                  col="Type", 
                  hue="Paid")

g.map(plt.scatter, 
      "Lifetime Post Total Reach", 
      "Lifetime Engaged Users",
      alpha=0.7)

g.add_legend()
```

```python
g = sns.FacetGrid(fb_data, 
                  row="Category",
                  col="Type", 
                  hue="Paid",
                  margin_titles=True)

# using regplot
g.map(sns.regplot, 
      "Post Month", 
      "Total Interactions",
      fit_reg=False)

g.add_legend()
plt.ylim(0, 2500)
```

```python
# specifying the each grid width

g = sns.FacetGrid(fb_data, 
                  row="Category",
                  col="Type", 
                  hue="Paid",
                  margin_titles=True,
                  gridspec_kws={"width_ratios": [5, 3, 2, 2]})

# using regplot
g.map(sns.regplot, 
      "Post Month", 
      "Total Interactions",
      fit_reg=False)

g.add_legend()
plt.ylim(0, 2500)
```

```python
h = {0: "red", 1:"green"}
# specifying the each grid width

g = sns.FacetGrid(fb_data, 
                  row="Category",
                  col="Type", 
                  hue="Paid",
                  margin_titles=True,
                  palette=h)

# using regplot
g.map(sns.regplot, 
      "Post Month", 
      "Total Interactions",
      fit_reg=False)

g.add_legend()
plt.ylim(0, 2500)
```

```python
# using different markers for value
h = {0: "red", 1:"green"}
i = {"marker":["^", "v"]}

# ^ - upper triangle
# v - inverted triangle
g = sns.FacetGrid(fb_data, 
                  row="Category",
                  col="Type", 
                  hue="Paid",
                  margin_titles=True,
                  palette=h,
                  hue_kws=i)

# using regplot
g.map(sns.regplot, 
      "Post Month", 
      "Total Interactions",
      fit_reg=False)

g.add_legend()
plt.ylim(0, 2500)
```

**Pair Grids**

Helps to analyze different relationships between different pairs of data that exists between datasets

```python
g = sns.PairGrid(fb_data, vars=["comment", "like", "share"])
g.map(plt.scatter)
```

```python
g = sns.PairGrid(fb_data, vars=["comment", "like", "share"])
g.map_offdiag(plt.scatter)
g.map_diag(sns.kdeplot)
```

```python
g = sns.PairGrid(fb_data, 
                 vars=["comment", "like", "share"],
                 hue="Type")
g.map_offdiag(plt.scatter)
g.map_diag(sns.kdeplot)
g.add_legend()
```

```python
g = sns.PairGrid(fb_data, 
                 vars=["comment", "like", "share"])

g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot)
g.map_lower(sns.regplot)
```

```python
g = sns.PairGrid(fb_data, 
                 x_vars=["comment", "like", "share"],
                 y_vars=["Post Month", "Post Weekday", "Post Hour"],
                 hue="Type")
g.map(plt.scatter)
g.add_legend()
```

## New Dataset

```python
bike_data = pd.read_csv("datasets/Bike-Sharing-Dataset/day.csv",
                        sep=r"\s*,\s*", 
                        engine="python")
bike_data.head()
```

```python
bike_data.describe()
```

```python
f, ax = plt.subplots(figsize=(15, 5))

sns.distplot(bike_data['cnt'], 
             bins=75, 
             kde=False, 
             rug=True)
```

```python
sns.jointplot(x="instant", 
              y="cnt",
             data=bike_data, 
            height= 8)
```

```python
sns.jointplot(x="mnth", 
              y="cnt",
             data=bike_data, 
            height= 8)
```

```python
sns.pairplot(bike_data,
             height=5,
             aspect=0.9,
            x_vars=["holiday", "weekday", "workingday"],
            y_vars=["cnt"],
            kind="reg",
            hue="season")
```

```python
sns.jointplot(x="season", 
              y="cnt",
             data=bike_data, 
            height= 8)
```

```python
sns.pairplot(bike_data,
             height=5,
             aspect=0.9,
            x_vars=["temp", "atemp","hum", "windspeed"],
            y_vars=["cnt"],
            hue="season")
```
